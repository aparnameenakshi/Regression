```python
# Import necessary libraries
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression, Lasso
from sklearn.metrics import mean_squared_error

# Sample dataset (replace this with your actual dataset)
years = np.array([2015, 2016, 2017, 2018, 2019, 2020, 2021])
prices = np.array([8000, 7500, 7000, 6500, 6000, 5500, 5000])

# Reshape the data for scikit-learn
years = years.reshape(-1, 1)

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(years, prices, test_size=0.2, random_state=42)

# Linear Regression
linear_reg_model = LinearRegression()
linear_reg_model.fit(X_train, y_train)
linear_reg_predictions = linear_reg_model.predict(X_test)

# Lasso Regression
lasso_reg_model = Lasso(alpha=0.01)  # You can adjust the alpha parameter
lasso_reg_model.fit(X_train, y_train)
lasso_reg_predictions = lasso_reg_model.predict(X_test)

# Evaluate models using Mean Squared Error
linear_reg_mse = mean_squared_error(y_test, linear_reg_predictions)
lasso_reg_mse = mean_squared_error(y_test, lasso_reg_predictions)

# Print Mean Squared Errors
print(f"Linear Regression MSE: {linear_reg_mse}")
print(f"Lasso Regression MSE: {lasso_reg_mse}")

# Predict the price for a 2022 model
year_2022 = np.array([[2022]])
linear_reg_2022_price = linear_reg_model.predict(year_2022)
lasso_reg_2022_price = lasso_reg_model.predict(year_2022)

print(f"Predicted price for a 2022 model using Linear Regression: {linear_reg_2022_price[0]}")
print(f"Predicted price for a 2022 model using Lasso Regression: {lasso_reg_2022_price[0]}")

# Plotting the results
plt.scatter(years, prices, color='black', label='Actual Prices')
plt.plot(X_test, linear_reg_predictions, color='blue', linewidth=3, label='Linear Regression')
plt.plot(X_test, lasso_reg_predictions, color='red', linewidth=3, label='Lasso Regression')
plt.scatter(year_2022, linear_reg_2022_price, color='blue', marker='X', s=100)
plt.scatter(year_2022, lasso_reg_2022_price, color='red', marker='X', s=100)
plt.title('Linear and Lasso Regression for Hyundai Grand i10')
plt.xlabel('Year')
plt.ylabel('Price')
plt.legend()
plt.show()
```
